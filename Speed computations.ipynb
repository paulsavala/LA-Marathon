```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data/mens_all_splits_processed.csv')
```

```python
df.head()
```

Speeds will be computed in meters per second

```python
def speed_computation(distance_km, minutes):
    speed = 1000*distance_km / (60*minutes)
    return speed

speed_computation = np.vectorize(speed_computation)
```

```python
df['5k_speed'] = speed_computation(5, df['5k_split'])
```

```python
df['10k_speed'] = speed_computation(5, df['10k_split'] - df['5k_split'])
df['15k_speed'] = speed_computation(5, df['15k_split'] - df['10k_split'])
df['20k_speed'] = speed_computation(5, df['20k_split'] - df['15k_split'])
df['25k_speed'] = speed_computation(5, df['25k_split'] - df['20k_split'])
df['30k_speed'] = speed_computation(5, df['30k_split'] - df['25k_split'])
df['35k_speed'] = speed_computation(5, df['35k_split'] - df['30k_split'])
df['40k_speed'] = speed_computation(5, df['40k_split'] - df['35k_split'])
df['last_2k_speed'] = speed_computation(2.195, df['clock_time'] - df['40k_split'])
```

```python
df['10k_avg_speed'] = speed_computation(10, df['10k_split'])
df['15k_avg_speed'] = speed_computation(15, df['15k_split'])
df['20k_avg_speed'] = speed_computation(20, df['20k_split'])
df['25k_avg_speed'] = speed_computation(25, df['25k_split'])
df['30k_avg_speed'] = speed_computation(30, df['30k_split'])
df['35k_avg_speed'] = speed_computation(35, df['35k_split'])
df['40k_avg_speed'] = speed_computation(40, df['40k_split'])
df['final_avg_speed'] = speed_computation(42.195, df['net_time'])
```

```python
df.to_csv('data/mens_speed.csv', index=False)
```
